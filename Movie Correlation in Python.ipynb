```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts configuration of plots

# read in data

df = pd.read_csv(r'C:\Users\Lee Zong Han\Desktop\Portfolio Projects\Project 4 - Correlation in Python\movies.csv')
```

```python
# look at the data

df.head()
```

```python
# check for null values

df.isnull().sum()
```

```python
# check for zeros

if np.any(df == 0):
    print("The DataFrame contains zeros")
else:
    print("The DataFrame does not contain zeros")
```

```python
df
```

```python
# drop rows with null values
df = df.dropna()
df
```

```python
# check data types
df.dtypes
```

```python
# change data types of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
# extract the year from the date column and create a new column

df["yearcorrect"] = df["released"].str.extract(r"(\d{4})")
df
```

```python
# check if original year column is same as released date

if df["year"].equals(df["yearcorrect"]):
    print("Years matching")
else:
    print("Years not matching")
```

```python
# sort by gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# change diplay to show all rows

pd.set_option('display.max_rows', None)
```

```python
# check if there any duplicates to drop

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df
```

```python
# hypothesis
# budget high correlation
# company high correlation
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Gross Earnings vs Budget')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color" : "blue"})
```

```python
# check for outliers

df.boxplot(column=['gross'])
```

```python
# look at correlation numbers

df.corr() #pearson (default), kendall, spearman
```

```python
# visualisation of correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# look at company correlation

df.head()
```

```python
df_numerised = df

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes
        
df_numerised
```

```python
df
```

```python
# visualisation of correlation matrix

correlation_matrix = df_numerised.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerised.corr()
```

```python
corr_mat = df_numerised.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
